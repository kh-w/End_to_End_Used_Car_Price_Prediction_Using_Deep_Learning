```python
import requests
import json
from bs4 import BeautifulSoup
from pathlib import Path
from time import sleep
import numpy as np
import datetime
import os
import re  # used later for string parsing

os.chdir('/Users/kh/Documents/Python Projects/used_car_pricing')

headers = {'User-Agent': 'Mozilla/5.0'}
now = str(datetime.datetime.now().year)+str('0'+str(datetime.datetime.now().month))[-2:]+str('0'+str(datetime.datetime.now().day))[-2:]
make = 'Kia'

f = open('linkvector.txt', 'r')
url = f.readlines()
f.close()

if not os.path.exists('cars'):
        os.makedirs('cars')
        
data_folder = Path("cars/")

for i in range(0,len(url)):

    singleurl = url[i][:-1]
    
    #sleep(np.random.choice([1,2,3,4,5])+3)

    response = requests.get(singleurl, headers=headers)
    soup = BeautifulSoup(response.content, 'lxml')
    
    print(str(i)+' of '+str(len(url)))
    print(singleurl)
    
    try:
        data = json.loads(soup.find('script',{'type':'application/ld+json'}).contents[0][9:-5])
    except:
        data = 'Unable to get data'
        
    file = open(data_folder / (make+now+"car"+str(i)+".txt"),"w") 
    file.write(str(data).replace("'",'"'))
    file.close()
```
